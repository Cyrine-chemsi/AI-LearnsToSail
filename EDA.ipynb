```python
def summary(data):
    # Get stats
    output_data = data.describe().transpose()
    # Count of NANs
    output_data['Count_NaNs'] = (data.shape[0] - output_data['count'])/ data.shape[0] * 100   
    # Count unique values
    output_data['Count_Distincts'] = data.apply(lambda x: len(pd.unique(x)), axis=0)  
    return output_data
summary(df)
```

#### Handling Duplicated Values

```python
# Get duplicated rows
duplicates = df.duplicated()

# Count the number of duplicated rows
print(duplicates.value_counts())

# Drop duplicates
df.drop_duplicates(keep = 'first', inplace = True)
```

#### Handling Missing Data

```python
#Imputing those 5 missing values in the datetime column

# 1 second timedelta
second = dt.timedelta(0,1)
# Populating null missing  in the 'DateTime' field
time_mask = df['DateTime'].isnull()
# Use the last valid DateTime
df['DateTime'] = df['DateTime'].fillna(method = 'bfill')
# Increment by one second that value
df.loc[time_mask, 'DateTime'] += second
```

```python
# Drop rows with more than half of the features missing
df.dropna(axis=0, thresh=len(df.columns)/2, inplace=True)

#Imputing missing values in the rest of the columns
df.interpolate(method='bfill', axis=0, limit=None, inplace = True)
```

Correlation analysis is one of those important checks that need to be performed during various stages of a project like data analysis, before and after feature transformations, feature engineering, and feature selection.

Correlation is a statistical measure that indicates the extent to which two or more variables move together. A positive correlation indicates that the variables increase or decrease together. A negative correlation indicates that if one variable increases, the other decreases, and vice versa. 

Below we represent the correlation matrix of all the input features: 

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(20, 20))
    ax = sns.heatmap(corr, annot=True, mask=mask, vmax=.5, vmin=-.5, square=True, cmap="YlGnBu")
    
```

The correlation matrix above for the numeric features indicates a high correlation between some of the features, for example SoG, SoS, AvgSoS and VMG seem highly correlated. Also TWA, AWA and other combinations. This indicates a possible problem of multicollinearity and the need for further investigation.

Multicollinearity is a situation where two or more predictors are highly linearly related. In general, an absolute correlation coefficient of > 0.7 among two or more predictors indicates the presence of multicollinearity.

Variance inflation factor (VIF) helps a formal detection-tolerance for multicollinearity. The cutoff to detect multicollinearity is VIF > 10.

About VIF Calculation : 

Assume we have a list of exogenous variable [X1, X2, X3, X4]. To calculate the VIF, we regress each of them against other exogenous variables. For instance, the regression model for X3 is defined as:  X3 ~ X1 + X2 + X4
And then we extract the R-squared from the model to calculate: VIF = 1 / (1 - R-squared)

   



```python
cols = list(df.columns)
cols.remove('Tacking')
cols.remove('DateTime')
cols.remove('ModePilote')

scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df[cols]),columns = df[cols].columns, index=df[cols].index)

# vif analysis    
vif_data = pd.DataFrame({'feature':cols})
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(scaled_df[cols].values, i) for i in range(len(cols))]
  
vif_data
```

```python
cols_to_keep = list(vif_data[vif_data['VIF']<=10].feature)
cols_to_export = cols_to_keep + ['Tacking'] + ['DateTime']
cols_to_export
```

```python
corr = df[cols_to_keep].corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(20, 20))
    ax = sns.heatmap(corr, annot=True, mask=mask, vmax=.7, vmin=-.7, square=True, cmap="YlGnBu")
```

As you can see in the above correlation matrix; after using VIF to remove columns with high multicollinearity, we no longer see highly correlated variables.

#### Check Data Balance 

```python
# Tacking distribution plot 
tacking = df['Tacking'].values
ax = sns.countplot(x = tacking)
fig = plt.gcf()
fig.set_size_inches(10,8)
ax.set_ylim(top=200000)
for p in ax.patches:
    ax.annotate('{:.2f}%'.format(100*p.get_height()/len(tacking)), (p.get_x()+ 0.333, p.get_height()+10000))

plt.xticks([0,1], ['No Tacking - 0', 'Tacking - 1'], rotation = 'horizontal')
plt.title('Distribution of Tacking Targets')
plt.xlabel('Tacking Labels', fontsize = 20)
plt.ylabel('Counts', fontsize = 20)
plt.show()
```

Our Dataset is imbalanced. This is a scenario where the number of observations belonging to one class is significantly lower than those belonging to the other classes.

In this situation, the predictive model developed using conventional machine learning algorithms could be biased and inaccurate.

This happens because Machine Learning Algorithms are usually designed to improve accuracy by reducing the error. Thus, they do not take into account the class distribution / proportion or balance of classes.
